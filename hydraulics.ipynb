```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Load the .npz file (assuming it's already loaded as `data`)
data = np.load('/content/CWRU_48k_load_1_CNN_data.npz')
X_data = data['data']
y_labels = data['labels']

# Check if the labels are strings and encode them if necessary
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_labels = label_encoder.fit_transform(y_labels)  # Convert labels to integers

# Normalize the data (scaling pixel values between 0 and 1)
X_data = X_data / 255.0  # Assuming pixel values between 0-255

# Reshape data for CNN: (samples, height, width, channels)
X_data = X_data.reshape(-1, 32, 32, 1)  # Assuming 32x32 images, and 1 channel (grayscale)

# Split data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_data, y_labels, test_size=0.2, random_state=42)

# Convert labels to one-hot encoding
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Create a CNN model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(y_train.shape[1], activation='softmax')  # Output layer
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Fit the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc * 100:.2f}%")

# Plot the training and validation accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Test Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')
plt.show()

# Plot the training and validation loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper left')
plt.show()
```

```python
import scipy.io

def load_and_view_mat(file_path):
    """
    Function to load and view contents of a .mat file.

    :param file_path: str, path to the .mat file
    """
    # Load the .mat file
    mat_data = scipy.io.loadmat(file_path)

    # Display the keys in the loaded .mat file (i.e., the variable names)
    print("Keys in the .mat file:", mat_data.keys())

    # Iterate through and display each variable's shape and content
    for key, value in mat_data.items():
        if key not in ['__header__', '__version__', '__globals__']:  # Skip metadata
            print(f"\nVariable: {key}")
            print(f"Shape: {value.shape}")
            print(f"Data:\n{value}")

    return mat_data

# Usage:
file_path = '/content/raw/Time_Normal_1_098.mat'  # Replace with your file path
mat_data = load_and_view_mat(file_path)
```

```python
import os
import random
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt

def load_and_plot_random_vibration_signal(folder_path='/content/raw/'):
    """
    This function randomly selects a .mat file from the specified folder,
    extracts vibration signal data (DE_time, FE_time), and plots the data.

    Parameters:
        folder_path (str): Path to the folder containing .mat files.
    """
    # List all .mat files in the folder
    mat_files = [f for f in os.listdir(folder_path) if f.endswith('.mat')]

    # Randomly select a file
    random_file = random.choice(mat_files)

    # Load the .mat file
    file_path = os.path.join(folder_path, random_file)
    mat_data = sio.loadmat(file_path)

    # Check available variables in the .mat file
    print(f"Variables in {random_file}: {mat_data.keys()}")

    # Search for keys that contain 'DE_time' and 'FE_time'
    de_key = None
    fe_key = None

    for key in mat_data.keys():
        if 'DE_time' in key:
            de_key = key
        elif 'FE_time' in key:
            fe_key = key

    # Extract the relevant data if available
    if de_key and fe_key:
        X_DE_time = mat_data[de_key].flatten()  # Flatten to make it 1D
        X_FE_time = mat_data[fe_key].flatten()  # Flatten to make it 1D

        # Print out basic information
        print(f"Loaded data from: {random_file}")
        print(f"Shape of {de_key}: {X_DE_time.shape}")
        print(f"Shape of {fe_key}: {X_FE_time.shape}")

        # Plot the vibration signals
        plt.figure(figsize=(10, 5))
        plt.subplot(2, 1, 1)
        plt.plot(X_DE_time)
        plt.title(f'{random_file} - {de_key} Vibration Signal')

        plt.subplot(2, 1, 2)
        plt.plot(X_FE_time)
        plt.title(f'{random_file} - {fe_key} Vibration Signal')

        plt.tight_layout()
        plt.show()
    else:
        print(f"Could not find relevant variables (DE_time, FE_time) in {random_file}")
```

```python
import scipy.io
import os
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

# Function to extract features from a time-domain signal
def extract_features(signal):
    features = {}
    features['max'] = np.max(signal)
    features['min'] = np.min(signal)
    features['mean'] = np.mean(signal)
    features['std'] = np.std(signal)
    features['rms'] = np.sqrt(np.mean(np.square(signal)))
    features['skewness'] = skew(signal)
    features['kurtosis'] = kurtosis(signal)
    features['crest'] = np.max(signal) / np.abs(np.mean(signal)) if np.abs(np.mean(signal)) > 0 else np.nan
    features['form'] = np.max(signal) / np.mean(np.abs(signal)) if np.mean(np.abs(signal)) > 0 else np.nan
    return features

# Function to load .mat files and extract features
def process_mat_files(directory):
    all_features = []  # List to store features of all files
    for filename in os.listdir(directory):
        if filename.endswith('.mat'):
            filepath = os.path.join(directory, filename)
            # Load the .mat file
            mat_data = scipy.io.loadmat(filepath)

            # Find all keys in the mat file
            mat_keys = mat_data.keys()

            # Try to extract time-domain signals
            signal_keys = [key for key in mat_keys if 'time' in key.lower()]

            for key in signal_keys:
                # Extract the signal (ensure it is a numpy array)
                signal = np.array(mat_data[key]).flatten()  # Flatten if it's a 2D array

                # Extract features from the signal
                features = extract_features(signal)
                features['file_name'] = filename  # Add the file name to the features
                all_features.append(features)

    # Create a DataFrame with all extracted features
    features_df = pd.DataFrame(all_features)
    return features_df

# Define the directory where the .mat files are stored
directory = '/content/raw/'

# Extract features from all .mat files in the directory
features_df = process_mat_files(directory)

# Display the first few rows of the features dataframe
print(features_df.head())
```
